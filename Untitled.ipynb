```python
import os
import json
import torch
import numpy as np
import torchvision.transforms as transforms
from PIL import Image
from tqdm import tqdm
from sklearn.preprocessing import LabelEncoder
from torch.utils.data import Dataset, DataLoader
import torch.nn as nn
import torch.optim as optim
```

```python
annotations_dirs = ["dataset/training_data/annotations", "dataset/testing_data/annotations"]
image_dirs = ["dataset/training_data/images", "dataset/testing_data/images"]

def load_layout_data(annotations_dirs):
    """
    Load bounding box information from JSON annotations.
    """
    layout_data = []
    for directory in annotations_dirs:
        for filename in tqdm(os.listdir(directory)):
            if filename.endswith(".json"):
                file_path = os.path.join(directory, filename)
                with open(file_path, "r", encoding="utf-8") as f:
                    data = json.load(f)

                    # Extract bounding boxes and labels
                    for item in data.get("form", []):
                        box = item.get("box", [])
                        label = item.get("label", "")
                        text = item.get("text", "")

                        if box:
                            layout_data.append({"box": box, "label": label, "text": text, "file": filename})

    return layout_data

# Load JSON data from both directories
layout_data = load_layout_data(annotations_dirs)

# Print sample data
print("Sample Layout Data:", layout_data[:5])
```

```python
def load_images(image_dirs, layout_data):
    """
    Load images corresponding to the annotation files.
    """
    images = []
    image_filenames = [item["file"].replace(".json", ".png") for item in layout_data]

    for directory in image_dirs:
        for img_file in os.listdir(directory):
            if img_file in image_filenames:
                img_path = os.path.join(directory, img_file)
                images.append(img_path)

    return images

# Load images
image_paths = load_images(image_dirs, layout_data)

# Print number of images loaded
print("Total Images Loaded:", len(image_paths))
```

```python
# Encode labels numerically
label_encoder = LabelEncoder()
labels = [item["label"] for item in layout_data]
encoded_labels = label_encoder.fit_transform(labels)

# Normalize bounding boxes (assuming 1024x1024 resolution)
boxes = np.array([item["box"] for item in layout_data]) / 1024.0
labels = np.array(encoded_labels)

# Define image transformations
transform = transforms.Compose([
    transforms.Resize((256, 256)),
    transforms.ToTensor()
])

# Custom Dataset Class
class ResumeLayoutDataset(Dataset):
    def __init__(self, images, boxes, labels, transform=None):
        self.images = images
        self.boxes = boxes
        self.labels = labels
        self.transform = transform

    def __len__(self):
        return len(self.images)

    def __getitem__(self, idx):
        image = Image.open(self.images[idx]).convert("RGB")
        if self.transform:
            image = self.transform(image)

        box = torch.tensor(self.boxes[idx], dtype=torch.float32)
        label = torch.tensor(self.labels[idx], dtype=torch.long)

        return image, box, label

# Create dataset and dataloader
dataset = ResumeLayoutDataset(image_paths, boxes, labels, transform)
dataloader = DataLoader(dataset, batch_size=16, shuffle=True)
```

```python
class LayoutGAN(nn.Module):
    def __init__(self, input_dim=4, hidden_dim=128, output_dim=4):
        super(LayoutGAN, self).__init__()
        self.fc1 = nn.Linear(input_dim, hidden_dim)
        self.fc2 = nn.Linear(hidden_dim, hidden_dim)
        self.fc3 = nn.Linear(hidden_dim, output_dim)

    def forward(self, x):
        x = torch.relu(self.fc1(x))
        x = torch.relu(self.fc2(x))
        x = self.fc3(x)
        return x

# Initialize model
model = LayoutGAN()
```

```python
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# Training loop
epochs = 10
for epoch in range(epochs):
    total_loss = 0
    for images, boxes, labels in dataloader:
        optimizer.zero_grad()
        outputs = model(boxes)
        loss = criterion(outputs, boxes)  # Training model to predict box coordinates
        loss.backward()
        optimizer.step()
        total_loss += loss.item()
    
    print(f"Epoch {epoch+1}/{epochs}, Loss: {total_loss/len(dataloader)}")

# Save the trained model
torch.save(model.state_dict(), "layoutgan_resume.pth")
```
